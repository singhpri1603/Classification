```python
class node:
    
    def __init__(self,col=-1,value=None,pred=None,lt=None,rt=None,status="running",gini=-1,size=-1,splits=[-1,-1,-1,-1]):
        self.col=col
        self.val=value
        #left tree will be matched value for string and less than for continuous
        self.left_tree=lt
        self.right_tree=rt
        self.answer=pred
        self.status=status
        self.gini=gini
        self.sample_size=size
        self.splits=splits
        
    

    def printnode(self):
        print [self.col,self.val,self.status,self.answer,self.gini,self.sample_size,self.splits]
        
    def getnode(self):
        if self.status=='finished':
            return [self.col,self.val,self.answer,"NA",self.sample_size,self.splits]
        else:
            return [self.col,self.val,"NA",self.gini,self.sample_size,self.splits]
    
pq=node(col=1,value=.22)
pq.printnode()
rs=node(col=1,value=.23,lt=pq)
rs.left_tree.printnode()
```

```python
def measures(l_p):
    a = 0
    b = 0
    c = 0
    d = 0
    for x in range(0,len(l_p)):
        if int(l_p[x][1]) == int(l_p[x][0]):
            #correct=correct+1
            if l_p[x][1] == 0:
                d = d+1
            else:
                a = a+1
        else:
            if l_p[x][1] == 0:
                b = b+1
            else:
                c = c+1

    
    accuracy = ((a+d)/float(a+b+c+d))*100
    if (a+c)==0:
        precision = 'denominator_0'
    else:
        precision = a/float(a+c)
    if (a+b)==0:
        #precision = 'denominator_0'
        recall = 'denominator_0'
    else:
        recall = a/float(a+b)
    if ((2*a)+b+c)==0:
        fmeasure = 'denominator_0'
    else: 
        fmeasure = (2*a)/float((2*a)+b+c)

    print "accuracy: " + str(accuracy)
    print "precision: " + str(precision)
    print "recall: " + str(recall)
    print "fmeasure: " + str(fmeasure)
    print ""
    return accuracy, precision, recall, fmeasure
```

```python
def fin_acc(l_p):
  siz=len(l_p)
  match=0
  for i in range(0,len(l_p)):
    if int(l_p[i][0])==int(l_p[i][1]):
      match=match+1
  acc=1.0*match/siz
  return acc
```

```python
import networkx as nx
import itertools as IT
import matplotlib.pyplot as plt

class BinaryTree(object):
   def __init__(self, data,name):
      self.data = data
      self.right = None
      self.left = None
      self.name = name
   def edgelist(self, counter = IT.count().next):
       self.name = counter() if self.name is None else self.name
       for node in (self.left, self.right):       
           if node:
               node.name = counter() if node.name is None else node.name
               yield (self.name, node.name)
       for node in (self.left, self.right):
           if node:
               for n in node.edgelist(counter):
                   yield n

tree = [BinaryTree(i,i) for i in range(5)]        

tree[0].left = tree[1]
tree[0].right = tree[2]
tree[2].left = tree[3]
tree[2].right = tree[4]

edgelist = list(tree[0].edgelist())
print(edgelist)   

G = nx.Graph(edgelist)
nx.draw_spectral(G)
plt.show()
```
